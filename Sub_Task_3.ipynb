```python
# Lower casing of the text

df['Article_text'] = df['Article_text'].str.lower()
```

```python
# Removing punctutation out of the data

import string
from string import punctuation
exclude = string.punctuation

def remove_punc1(text):
    return text.translate(str.maketrans('','',exclude))
```

```python
df['Article_text'] = df['Article_text'].apply(remove_punc1)
df['Article_text'].head()
```

```python
# Tokenization

import nltk
nltk.download('punkt')

from nltk.tokenize import word_tokenize

df['text_tokens'] = df['Article_text'].apply(word_tokenize) # inserting a new column as text token for further analysis
```

```python
for index, row in df.iterrows():
  article_index = row['News_ID']
  arti_text = row['text_tokens']
  print(article_index,arti_text)
```

```python
# Removing Stopwords

import nltk
nltk.download('stopwords')

from nltk.corpus import stopwords

filter_words = []
filter_text = []

for index, row in df.iterrows():

    article_index = row['News_ID']
    arti_text = row['text_tokens']

    stop_words = set(stopwords.words('english'))    # It will store the entire stop words list
    filtered_words = [word for word in arti_text if word not in stop_words and word.isalpha()]  # removing stopwords and numeric values in filtered words
    filtered_text = ' '.join([word for word in filtered_words])         # join the filter word to reform the text as filter text
    filter_words.append(filtered_words)        # appending each articles filter words in list
    filter_text.append(filtered_text)          # appending each articles filter text in list


df['Filter_words'] = pd.Series(filter_words)  # Inserting new column as filter word to perform word frequency analysis
df['Filter_text'] = pd.Series(filter_text)    # Inserting new column as filter text to perform sentiment analysis
```

# Word Frequency Analysis

```python
import nltk
word_freq_count = nltk.FreqDist(df['Filter_words'][0])  # this will calculate the frequency of word in text
word_freq_count
```

```python
print(word_freq_count.most_common(20))  # this function will show top most frequent words
```

```python
# Other method to perform word frequency analysis but the results are same

from collections import Counter

word_counts = Counter(df['Filter_words'][0])           # this will calculate the frequency of word in text
most_common_words = word_counts.most_common(10)        # this function will show the top most frequent words
print(most_common_words)
```

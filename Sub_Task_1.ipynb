```python
# Importing the necessary library to perform webscraping
import pandas as pd           # It is the core library of python to perform data analysis  
import requests               # Its library to send request to the server or website to fetch data out of it.
from bs4 import BeautifulSoup # Its one of the library of python used for doing web scrapping
```

```python
# Initially we create http request using request library to fetch data from the website or server

headers = {'User-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/110.0.0.0 Safari/537.36'}   # it is generally used to show server believe that its request coming from some browser otherwise sometimes server consider that it's a request coming from some malcious bot!
webpage = requests.get('https://www.indiatoday.in/india', headers=headers).text   # its response we get in return of our request and store it in webpage object
```

```python
webpage # this is the response we get in return of our request, its basically the structure or content of webpage, from this we can scrap data for our analysis
```

```python
soup = BeautifulSoup(webpage, 'lxml') # here i create soup object using beautifulsoup, in this i m passing an html parser that will run around the entire html content.
```

```python
# Now we can extract any part of the webpage using soup object using different beautifulSoup's methods

print(soup.prettify()) #it one of the method in beautifulsoup that gives us the content of webpage in correct order so that we can able to understand what actually written in html.
```

```python
# here i m storing all the article tag in list to variable Articles

Articles = soup.find_all('article', class_ = 'B1S3_story__card__A_fhi')
len(Articles)
```

```python
# here i m iterating through Articles list to fetch text out of the relevent tags and store in the respective list.

News_source = []
Article_title = [] 
Article_headline = []
Image_source = []
News_loc = []

for i in Articles:
    News_source.append("https://www.indiatoday.in" + i.find('a').get('href')) # Extracting the respective link of each article present on webpage to get the entire text of webarticles
    Article_title.append(i.find('h2').text)      # Extracting the title of each articles on webpage
    Article_headline.append(i.find('p').text)    # Extracting the headline of each articles on webpage
    Image_source.append(i.find('img').get('src'))  # Extracting the image source for each article on webpage 
    News_loc.append(i.find('h4').text)            # Extracting the news location of each articles on webpage


# len(news_link), len(article_headline), len(article_title), len(image_source), len(newstype)

# here i m creating a dataframe using dictionary
import numpy as np
d = {'News_loc':News_loc, 'Article_title':Article_title, 'Article_headline':Article_headline, 'Image_source':Image_source, 'News_Source':News_source}    
df = pd.DataFrame(d)
df.insert(0,'News_ID',pd.Series(np.arange(101,121)))
df
# so this is basically the scraped data
```

```python
# here i m hitting the link of the respective Web article in the dataframe through iteration to extract entire text data of  particular article out of it.  

extract_text = []
for index, row in df.iterrows():

    url_index = row['News_ID']
    url = row['News_Source']

    response = requests.get(url).text

    soup = BeautifulSoup(response, 'lxml') 

    # Filter the paragraphs tags, excluding those <p> containing 'Listen to Story'
    p_filter = [p for p in soup.find_all('p') if 'Listen to Story' not in p.get_text()]
        
    # Extract text from each <p> tag and join them into a single string
    full_text = ' '.join([p.get_text(strip=True) for p in p_filter if p.get_text(strip=True)])
        
    # Append the full text to the extract_text list
    extract_text.append(full_text)

# here I m creating the new column for article text in a dataframe
df['Article_text'] = pd.Series(extract_text)
df
```

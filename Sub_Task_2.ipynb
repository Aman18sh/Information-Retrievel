# Basic Text Processing

```python
# Removing punctutation out of the text data

import string
from string import punctuation
exclude = string.punctuation

def remove_punc1(text):             # function to remove punctuation from the text
    return text.translate(str.maketrans('','',exclude))
```

```python
df['Article_text'] = df['Article_text'].apply(remove_punc1)
df['Article_text'].head()
```

# Tokenization

```python
# Tokenization - breaking text in to small parts, it could be word or character

import nltk
nltk.download('punkt')

from nltk.tokenize import word_tokenize # importing tokenize function from tokenize module to perform tokenization

df['Article_text'] = df['Article_text'].apply(word_tokenize)  # first tokenize and assign to the same column
```

```python
df['Article_text'][0][0:3]
```

# Stemming

```python
# Stemming - reducing the word in to its roots form

from nltk.stem import PorterStemmer

stemmer = PorterStemmer() # Creating an object for class stemmer

df['Article_text'] = df['Article_text'].apply(lambda tokens: [stemmer.stem(token) for token in tokens]) # first did stemming and assign to the same colum
df['Article_text']
```

# Lemmatization

```python
# Lemmatization - reducing the word in to base

from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')

lemmatizer = WordNetLemmatizer() # creating an object for lemmatizer

df['Article_text'] = df['Article_text'].apply(lambda tokens: [lemmatizer.lemmatize(token) for token in tokens])  # first did lemmatization and assign to the same column
df['Article_text']
```

```python
# showing each each row and its tokens

for index, row in df.iterrows():
  article_index = row['News_ID']
  arti_text = row['Article_text']
  print(article_index,arti_text)
```

# Stopwords Removal

```python
# Removing Stopwords out of the text tokens

import nltk
nltk.download('stopwords')

from nltk.corpus import stopwords

filter_text = []
filter_words = []

for index, row in df.iterrows(): # Iterating through each row it will return the index and raw value, here in data i append the row index as news id during the first task which i consider as row index here

    article_index = row['News_ID']
    arti_text = row['Article_text']

    stop_words = set(stopwords.words('english'))    # It will store the entire stopwords list
    filtered_words = [word for word in arti_text if word not in stop_words and word.isalpha()] # removing stopwords amd numeric values
    filtered_text = ' '.join([word for word in filtered_words])   # here I m joining the words to form the entire text again as filter text

    filter_text.append(filtered_text)      # appending the filter text in a list
    filter_words.append(filtered_words)    # append the filter words in a list

df['Filter_text'] = pd.Series(filter_text) # inserting the new column as filter text
```

```python
df.head()
```

# Text Correction

```python
# Text Correction

from textblob import TextBlob

def text_correction(text):
    return str(TextBlob(text).correct())

df['Filter_text'] = df['Filter_text'].apply(text_correction)
```

```python
df['Filter_text'][0]
```

# Text Vectorization

```python
# Text Vectorization

from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer()

Vect_Articles = tfidf.fit_transform(df['Filter_text']).toarray()  # converting the text into numbers using tf-idf
Vect_Articles
```

# Standardizing Features

```python
# Standardising Features

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_articles = scaler.fit_transform(Vect_Articles)    # standardising features to get particular range so that there is not much difference in the features
scaled_articles
```

```python
numer_col = pd.DataFrame(scaled_articles)
numer_col.head()
```

# Outlier Detections

```python
# Removing Outliers

# here i m using interquartile range method to remove outliers

Q1 = np.percentile(scaled_articles, 25, axis=0)
Q3 = np.percentile(scaled_articles, 75, axis=0)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR #lower threshold
upper_bound = Q3 + 1.5 * IQR #upper threshold

outliers = np.any((scaled_articles < lower_bound) | (scaled_articles > upper_bound), axis=1)
X_filtered = scaled_articles[~outliers]
X_filtered
outliers
# It will showing ouliers in all the rows due to small dataset or very narrow inter quartile range
```
